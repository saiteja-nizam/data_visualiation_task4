###### Department wise codekata performence (pie chart)

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mean_score=df3.groupby('Department').mean()['CodeKata Score']
print(mean_score)
department=['CSE','ECE','EEE']
plt.pie(mean_score,labels=department,autopct='%0.2f%%',radius=1.5,shadow=True)
plt.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
toppers=df3.sort_values(['Department','CodeKata Score'],ascending=(0,0)).groupby('Department')['Name','CodeKata Score','Department'].head(3)
toppers
```

```python
x=toppers['Name'].iloc[0:3]
y=toppers['CodeKata Score'].iloc[0:3]
a=toppers['Name'].iloc[3:6]
b=toppers['CodeKata Score'].iloc[3:6]
c=toppers['Name'].iloc[6:9]
d=toppers['CodeKata Score'].iloc[6:9]
plt.xlabel('CodeKata Score')
plt.ylabel('Student Name')
plt.title('Departmentwise toppers')
plt.barh(x,y,label='EEE')
plt.barh(a,b,label='ECE')
plt.barh(c,d,label='CSE')
plt.grid()
plt.legend()
plt.show()
```
